```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#on change l'index par defaut pour que ça devienne Date et que pandas l'interprète comme une date
bitcoin= pd.read_csv(r"C:/Users/Aboubakr Ben Bach/Desktop/ML/BTC-EUR.csv", index_col='Date', parse_dates=True)
bitcoin.head()
```

```python
#visualiser
bitcoin['Close'].plot(figsize=(9,6))
plt.show()
```

```python
bitcoin.index
#l'evolution du bitcoin en 2019
bitcoin['2019-09']['Close'].plot()
```

```python
#l'evolution du bitcoin entre 2017 et 2019
bitcoin.loc['2017':'2019','Close'].plot()
```

```python
#resample: pour regrouper les données selon
bitcoin['2019']['Close'].resample('M').plot()
plt.show()
```

```python
#on change l'index par defaut pour que ça devienne Date et que pandas l'interprète comme une date
ethereum= pd.read_csv(r"C:/Users/Aboubakr Ben Bach/Desktop/ML/ETH-EUR.csv", index_col='Date', parse_dates=True)
ethereum.head()
```
